```python
# Grafico de u(x,y)=0
def ReU(x):
  return (1+(x+1)**2)**(1/2)

x = np.linspace(-4,2,100)
y = ReU(x)
```

```python
import pandas as pd
import plotly.graph_objects as go
```

```python

def wU(x,y):
  return x**2 + 2*x + 2 - y**2
def wV(x,y):
  return 2*x*y + 2*y

Y = np.linspace(-2,2,100)
X = np.linspace(-4,2,100)

Y2=ReU(X)

Z = np.zeros([100,100])
Z2 = np.zeros([100,100])
Y3=np.full(100,-1)
for i in range(100):
   Z[i]= wU(X,Y[i])
   Z2[i]=wV(X,Y[i])

```

```python
from sympy import *
x, y, z, U = symbols('x y z U')
U = x**(2)+2*x+2-y**(2)
U
```

```python
from plotly.subplots import make_subplots

fig = go.Figure(data=[go.Surface(
    colorscale="jet",
    x = X,
    y = Y,
    z = Z, surfacecolor = Z2)])

fig.add_scatter3d(x=X.flatten(),y=Y2.flatten(), z=np.zeros([100]), mode='lines',line=dict(color='Black', width=6))
fig.add_scatter3d(x=X.flatten(),y=-Y2.flatten(), z=np.zeros([100]),mode='lines',line=dict(color='Black', width=6))
fig.add_scatter3d(x=Y3,y=Y.flatten(), z=np.zeros([100]),mode='lines',line=dict(color='#0000FF', width=6))
fig.add_scatter3d(x=X,y=np.zeros([100]), z=np.zeros([100]),mode='lines',line=dict(color='#0000FF', width=6))
fig.update_layout(title=f'U(x,y)={U}', showlegend=False)

fig.show()

```
